```python

import pandas as pd

train = pd.read_csv('/Users/kshitijsrivastava/Downloads/Case study 2/train.csv')
test = pd.read_csv('/Users/kshitijsrivastava/Downloads/Case study 2/test.csv')
train.head(5)
```

```python
NAs = pd.concat([train.isnull().sum()], axis=1 , keys=['train'])
NAs[NAs.sum(axis=1) > 0]
```

```python
train["LotFrontage"] = train["LotFrontage"].fillna(train["LotFrontage"].mean())
```

```python
train["MasVnrArea"] = train["MasVnrArea"].fillna(train["MasVnrArea"].mean())
```

```python
train["GarageYrBlt"] = train["GarageYrBlt"].fillna(train["GarageYrBlt"].mean())
```

```python
display(train)
```

```python
NAs = pd.concat([train.isnull().sum()], axis=1 , keys=['train'])
NAs[NAs.sum(axis=1) > 0]
```

```python
for col in train.dtypes[train.dtypes == "object"].index:\
for_dummy = train.pop(col)
train = pd.concat([train, pd.get_dummies(for_dummy, prefix=col)],axis=1)
train.head
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
labels = np.array(train['SalePrice'])
# Remove the labels from the features
# axis 1 refers to the columns
features= train.drop('SalePrice', axis = 1)
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(features, labels, test_size = 0.1, random_state = 100000)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier()
# Train the model on training data
rf.fit(X_train,Y_train);
```

```python
y_predict = rf.predict(X_test)
```

```python
import math
MSE = np.square(np.subtract(Y_test,y_predict)).mean() 
 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)


```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test, y_predict)
```

```python
from sklearn import datasets, linear_model, metrics
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
y_predict = reg.predict(X_test)
```

```python
import math
MSE1 = np.square(np.subtract(Y_test,y_predict)).mean() 
 
RMSE1 = math.sqrt(MSE1)
print("Root Mean Square Error:\n")
print(RMSE1)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test, y_predict)
```

```python
from sklearn.tree import DecisionTreeRegressor 
regressor = DecisionTreeRegressor(random_state = 100000) 
  
# fit the regressor with X and Y data
regressor.fit(X_train, Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test, y_pred)
```

```python
from sklearn.svm import SVR
# most important SVR parameter is Kernel type. It can be #linear,polynomial or gaussian SVR. We have a non-linear condition #so we can select polynomial or gaussian but here we select RBF(a #gaussian type) kernel.
regressor = SVR(kernel='rbf')
regressor.fit(X_train,Y_train)
#5 Predicting a new result
y_pred = regressor.predict(X_test)
```
